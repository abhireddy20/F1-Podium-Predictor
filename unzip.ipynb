```python
import zipfile
import os
 
# ── Change this to the name of your downloaded Kaggle zip file ──
ZIP_FILE = "archive.zip"  # rename if your file is called something different
 
OUTPUT_DIR = "data"
 
os.makedirs(OUTPUT_DIR, exist_ok=True)
 
print(f"Extracting '{ZIP_FILE}' into '{OUTPUT_DIR}/'...")
 
with zipfile.ZipFile(ZIP_FILE, "r") as z:
    z.extractall(OUTPUT_DIR)
 
print("✅ Done! Files extracted:")
for f in sorted(os.listdir(OUTPUT_DIR)):
    print(f"   data/{f}")
```
